### 1.2 Implementación del modelo

Lo primero, implementa una función que devuelva los valores de $\widehat{w}$ dados X e y. Si se da el caso de que el dataset tiene más features que observaciones, queremos transponerlo antes de realizar las operaciones. Lo hacemos porque, como vimos en clase, los resultados no van a ser buenos si el número de observaciones es menor.

```python
import numpy as np

def pesos_ols(X, y):
    # Implementa la fórmula matricial
    # TODO 1 Comprueba que las dimensiones son adecuadas y corrige si no es así
    print("Dimensiones de la matriz inicial",X.shape) 
    n_filas = X.shape[0]
    n_columnas = X.shape[1]
    if n_filas < n_columnas:
        X = X.T
        print("Dimensiones de la matriz corregida",X.shape) 
    # TODO 2 Prepara la matriz X para que tenga la forma adecuada para aplicar la fórmula 
    ones = np.ones((X.shape[0],1))
    print("Dimensiones matriz de unos",ones.shape)
    X_o = np.concatenate((ones, X), axis = 1)
    print("Dimensiones de la matriz concatenada con unos 'X_o'",X_o.shape)
    X_ot=X_o.T
    print("Dimensiones de la matriz traspuesta",X_ot.shape) 
    print("Dimensiones de matriz y",y.shape)  
    # TODO 3 Calcula los pesos
    w = np.linalg.inv(np.dot(X_ot, X_o)).dot(X_ot).dot(y)
    #print(w)
    # comprueba que los valores calculados son correctos
    #wt, _, _, _ = np.linalg.lstsq(X_o, y, rcond = None)
    #print(wt)
    return w
```

```python
X = np.array([[1710, 1262, 1786, 1717, 2198, 1362, 1694, 2090, 1774, 1077], 
              [2003, 1976, 2001, 1915, 2000, 1993, 2004, 1973, 1931, 1939]])
y = np.array([208500, 181500, 223500, 140000, 250000, 143000, 307000, 200000, 129900, 118000])
w = pesos_ols(X, y)
print(w)
```

### 1.4 Compara el método de mínimos cuadrados y el descenso del gradiente para obtener los coeficientes de una regresión lineal



## 1. Método de Mínimos Cuadrados Ordinarios (OLS)

### Características:
- **Enfoque Analítico**: OLS es un método cerrado que encuentra la solución exacta para los coeficientes usando la fórmula de la ecuación normal.
- **Fórmula**: La ecuación para calcular los coeficientes es:
$W'=(X^{T}X)^{-1}X^{T}y$
  donde $X$ es la matriz de características e $y$ es el vector de valores objetivo.
- **Requisitos de Datos**: El método asume que $X^{T}X$ es invertible. Si la matriz no es invertible (o es singular), lo que puede ocurrir con datos altamente correlacionados (colinealidad), la fórmula no se puede aplicar directamente.
- **Computacionalmente Intensivo**: Invertir una matriz puede ser costoso en términos computacionales, especialmente si la matriz es grande o si se trata de problemas de alta dimensionalidad (muchas características).
- **Adecuado para**: Problemas de regresión con una cantidad moderada de datos y características, donde se puede calcular de manera eficiente y se necesita una solución exacta.

## 2. Descenso del Gradiente

### Características:
- **Enfoque Iterativo**: Este método es una técnica numérica que encuentra la solución minimizando iterativamente la función de costo (como el error cuadrático medio) moviéndose en la dirección opuesta al gradiente.
- **Proceso**: Comienza con un conjunto inicial de coeficientes (pesos) y ajusta estos coeficientes en pasos sucesivos, proporcionales a la derivada parcial de la función de costo con respecto a cada coeficiente.
- **Tasa de Aprendizaje**: La magnitud de los pasos es controlada por un parámetro llamado tasa de aprendizaje. Si es muy grande, puede causar oscilaciones y no converger; si es muy pequeña, puede hacer que el proceso sea muy lento.
- **Computacionalmente Eficiente**: Particularmente útil para grandes conjuntos de datos y cuando el número de características es alto, ya que no requiere invertir matrices grandes.
- **Flexibilidad**: Puede ser utilizado en situaciones donde $X^{T}X$  no es invertible o donde los datos no caben en la memoria, ya que se puede aplicar a sublotes (mini-batches) de datos.
- **Adecuado para**: Problemas con grandes conjuntos de datos y alta dimensionalidad, y cuando se busca una solución aproximada que sea computacionalmente factible.

## Comparación General

- **Precisión**: OLS proporciona una solución exacta bajo la suposición de que $X^{T}X$  es invertible y no hay problemas numéricos. El descenso del gradiente puede aproximarse a la solución exacta, pero la precisión depende de factores como la tasa de aprendizaje y el número de iteraciones.
- **Computación**: OLS puede ser más costoso computacionalmente en problemas de alta dimensionalidad debido a la necesidad de calcular la inversa de una matriz grande. El descenso del gradiente es más adecuado para grandes conjuntos de datos debido a su naturaleza iterativa.
- **Robustez**: El descenso del gradiente es más robusto frente a problemas de colinealidad y a conjuntos de datos grandes que no caben en la memoria.

Vamos a resolver el mismo problema usando el descenso del gradiente en lugar de la solución analítica. El descenso del gradiente es un método de optimización iterativo que usa el operador gradiente.

Como es iterativo, necesitamos decirle cuándo tiene que parar y darnos la solución. En este caso, le vamos a decir que pare y nos devuelva la solución después de un número de iteraciones que le pasaremos como parámetro. La ecuación iterativa es.

$$w^{t+1} = w^t - \eta \cdot \nabla f(w)$$

donde f es nuestra función objetivo, y w es un vector. Para la función objetivo, vamos a usar el error RSS, que querremos minimizar.

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

y cuyo gradiente es:

$$\nabla RSS(w) = X^T(Xw^t-y)$$

TIP: Ten en cuenta que el RSS también lo puedes escribir como $\frac{1}{2}\sum_{n=1}^{N}[y_n-\hat{y_n}]^2$ donde $\hat{y_n} = X \hat{w}$ en cada paso.
